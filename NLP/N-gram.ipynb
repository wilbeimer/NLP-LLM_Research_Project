```python
file = open("Books/Pride_and_Prejudice_Jane_Austin.txt", "r")
corpus = file.read()
tokens = re.findall(r"[A-Za-z0-9']+|[.,!?;]", corpus.lower())
```

```python
trigrams = defaultdict(lambda: defaultdict(int))

for i in range(len(tokens) - 2):
    w1, w2, w3 = tokens[i], tokens[i+1], tokens[i+2]
    trigrams[(w1, w2)][w3] += 1
```

```python
model = {}

for context, next_words in trigrams.items():
    total = sum(next_words.values())
    model[context] = {word: count / total for word, count in next_words.items()}
```

```python
def generate_text(model, num_words=30):
    # pick a random starting pair (context)
    context = random.choice(list(model.keys()))
    w1, w2 = context
    output = [w1, w2]

    for _ in range(num_words):
        next_words = model.get((w1, w2), None)
        if not next_words:
            break  # no next word known for this pair

        # randomly choose next word based on probabilities
        next_word = random.choices(
            list(next_words.keys()), 
            weights=next_words.values()
        )[0]

        output.append(next_word)
        w1, w2 = w2, next_word  # move context forward

    return ' '.join(output)
```

```python
generate_text(model)
```
